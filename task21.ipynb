# **Univariate Analysis**

Uni means single and variate means variable. Hence univariate means the analysis of a single variable. In univariate analysis, we are trying to analyse the pattern present in a single variable.

Histogram: Histogram is used to plot the graphical representation of a numerical variable. The x-axis represents the numeric data and y-axis represents the respective count of the data.

```python
import matplotlib.pyplot as plt
plt.hist(data["churn"])           # Histogram plot using matplotlib
plt.xlabel("churn")
plt.ylabel("Count")
```

```python
# Assuming "churn" is a binary column where 0 represents not churned and 1 represents churned
churn_counts = data["churn"].value_counts()

# Display the counts
print("Not Churned:", churn_counts[0])
print("Churned:", churn_counts[1])
```

so people churned are 2037 and not churned are 7963

```python
import seaborn as sns
sns.histplot(data, x='churn')        # Histogram plot using seaborn
```

## **Box plot**: Box plot is created to get the summary of variable ie. minimum, first quartile, median, 3rd quartile and maximum.

```python
plt.boxplot(data["age"])        # Box plot using matplotlib
```

```python
sns.boxplot(data["age"])        # Box plot using seaborn
```

**Univariate Analysis for Categorical Variables**



```python
sns.countplot(data, x="gender")
```

```python
# Group the data by the `gender` column.
grouped = data.groupby("gender")

counts = grouped.size()

total_males = counts["Male"]
total_females = counts["Female"]

# Print the total number of male and female customers.
print(total_males)
print(total_females)
```

```python
# Bar graph using matplotlib
gender_counts = data['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Bar graph using seaborn
gender_counts = data['gender'].value_counts()
sns.barplot(data, x=gender_counts.index, y=gender_counts.values)
```

```python
# Pie chart using matplotlib
gender = data['gender'].value_counts()
plt.pie(gender.values, labels=gender.index, autopct='%1.1f%%')
plt.title("gender Distribution")
```

54.6% are males and 45.4% are females

# **BIVARIATE ANALYSIS**

Bivariate means the analysis of two variables. Using bivariate analysis we can find how well the variables are correlated. Bivariate analysis is of 3 types

Numerical variables Categorical variables Numerical & Categorical variable

Scatterplot: Scatterplot uses dots to represent the relationship between two numeric variables.

```python
# Scatter plot using Seaborn
sns.scatterplot(data, x="age", y="balance")
plt.title("Scatter Plot of age vs balance")
plt.xlabel("age")
plt.ylabel("balance")
```

```python
# Scatter plot using matplotlib
plt.scatter(data["age"], data["balance"])
plt.title("Scatter Plot of Age vs balance")
plt.xlabel("Age")
plt.ylabel("balance")
plt.show()
```


age between 30-40 is more dense for 150000-10000 balance

```python
# Scatter plot using matplotlib
plt.scatter(data["age"], data["churn"])
plt.title("Scatter Plot of Age vs churn")
plt.xlabel("Age")
plt.ylabel("churn")
plt.show()
```

```python
#estimated_salary
# Scatter plot using matplotlib
plt.scatter(data["age"], data["estimated_salary"])
plt.title("Scatter Plot of Age vs cestimated_salary")
plt.xlabel("Age")
plt.ylabel("estimated_salary")
plt.show()
```

again age 20-30 is more dense fro income 0-20000
