# **Univariate Analysis**

Uni means single and variate means variable. Hence univariate means the analysis of a single variable. In univariate analysis, we are trying to analyse the pattern present in a single variable.

Histogram: Histogram is used to plot the graphical representation of a numerical variable. The x-axis represents the numeric data and y-axis represents the respective count of the data.

```python
import matplotlib.pyplot as plt
plt.hist(data["Electric Range"])           # Histogram plot using matplotlib
plt.xlabel("Electric Range")
plt.ylabel("Count")
```

```python
import matplotlib.pyplot as plt
plt.hist(data["Base MSRP"])           # Histogram plot using matplotlib
plt.xlabel("Base MSRP")
plt.ylabel("Count")
```

```python
import matplotlib.pyplot as plt
plt.hist(data["Model Year"])           # Histogram plot using matplotlib
plt.xlabel("Model Year")
plt.ylabel("Count")
```

```python
import matplotlib.pyplot as plt
plt.hist(data["DOL Vehicle ID"])           # Histogram plot using matplotlib
plt.xlabel("DOL Vehicle IDr")
plt.ylabel("Count")
```

```python
import matplotlib.pyplot as plt
plt.hist(data["Postal Code"])           # Histogram plot using matplotlib
plt.xlabel("Postal Code")
plt.ylabel("Count")
```

```python
import matplotlib.pyplot as plt
plt.hist(data["2020 Census Tract"])           # Histogram plot using matplotlib
plt.xlabel("2020 Census Tract")
plt.ylabel("Count")
```

```python
# Pie chart using matplotlib
data = data.head(5)
Make = data['Make'].value_counts()
plt.pie(Make.values, labels=Make.index, autopct='%1.1f%%')
plt.title("Make Distribution")
```

```python
# Pie chart using matplotlib
data = data.head(5)
City = data['City'].value_counts()
plt.pie(City.values, labels=City.index, autopct='%1.1f%%')
plt.title("City Distribution")
```
